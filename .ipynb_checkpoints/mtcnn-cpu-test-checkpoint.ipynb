### 6700 개 이미지 시간 측정 결과

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
diffs = np.load(file="./cpu_result_times.npy")
g_diffs = np.load(file="./gpu_result_times.npy")
```

### MTCNN CPU/GPU 버전 시간 측정 (500ms 이상 제외)
<ul>
    <li><b><span style="color:lightgreen">초록색 CPU</span></b></li>
    <li><b><span style="color:red">빨간색 GPU</span></b></li>
</ul>

```python
plt.hist(diffs, bins=100, range=(diffs.min(), 500), color="green", alpha=0.5)
plt.hist(g_diffs, bins=100, range=(g_diffs.min(), 500), color="red", alpha=0.5)
plt.title("MTCNN CPU")
plt.xlabel("Milliseconds")
plt.ylabel("Frequency")
```

### 통계 수치 비교 및 500ms 이상 걸리는 케이스 갯수 및 비율

```python
from pandas import DataFrame, Series
data = {
    'CPU': diffs,
    'GPU': g_diffs
}
df = DataFrame(data)

num_gt_500 = sum([1 for i in diffs if i > 500])
g_num_gt_500 = sum([1 for i in g_diffs if i > 500])
df.describe().append(
    Series({
        'CPU':(num_gt_500, "%.2f%%" %(num_gt_500/len(diffs) * 100)), 
        'GPU':(g_num_gt_500, "%.2f%%" %(g_num_gt_500/len(diffs) * 100))
    }, name='500ms <')
)
```

### 이미지 사이즈 분포 (CPU)
<ul>
    <li><b><span style="color:lightgreen">500ms 이하 초록색</span> (95%)</b></li>
    <li><b><span style="color:red">500ms 이상 빨간색</span> (5%)</b></li>
</ul>

```python
sizes = np.load(file="sizes.npy")
THRESHOLD = 500

def size(select_lt_thres, times):
    return [int(select_lt_thres) if time < THRESHOLD else int(not select_lt_thres) for time in times]
def color():
    return ["lightgreen" if time < THRESHOLD else "red" for time in diffs]

plt.figure(figsize=(5, 7))
plt.scatter(sizes[:,1], sizes[:,0], s=size(True), c=color())
plt.scatter(sizes[:,1], sizes[:,0], s=size(False), c=color())
plt.xlabel("width")
plt.ylabel("height")
```

### GPU

```python
plt.figure(figsize=(5, 7))
plt.scatter(sizes[:,1], sizes[:,0]
            , s=[1 if time < THRESHOLD else 0 for time in g_diffs]
            , c=["lightgreen" if time < THRESHOLD else "red" for time in g_diffs])
plt.scatter(sizes[:,1], sizes[:,0]
            , s=[0 if time < THRESHOLD else 1 for time in g_diffs]
            , c=["lightgreen" if time < THRESHOLD else "red" for time in g_diffs])
plt.xlabel("width")
plt.ylabel("height")
```
